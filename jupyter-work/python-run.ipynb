```python
import requests
import json
from pyspark.context import SparkContext
from pyspark.sql.session import SparkSession
import pyspark.sql.functions as F
from datetime import datetime
from datetime import timedelta
from re import sub

today=datetime.now().strftime("%Y-%m-%d")
yesterday=(datetime.now()-timedelta(1)).strftime("%Y-%m-%d")

print(today,yesterday,"starting tasks at:",datetime.now())

pd.set_option('display.max_rows', 1000)
pd.options.display.float_format = '{:.2f}'.format

warehouse_location_path = '/home/jovyan/work/spark-warehouse'

warehouse_location = abspath(warehouse_location_path)

print(warehouse_location)

spark = SparkSession \
    .builder \
    .appName("shopee-category-search-extract") \
    .config("spark.sql.warehouse.dir", warehouse_location) \
    .config("spark.debug.maxToStringFields",200) \
    .config("spark.sql.debug.maxToStringFields",2000) \
    .config("spark.executor.memory", "4g") \
    .config("spark.driver.memory", "4g") \
    .config("spark.executor.extraJavaOptions","-Dio.netty.tryReflectionSetAccessible=true -Xms4096m") \
    .config("spark.driver.extraJavaOptions","-Dio.netty.tryReflectionSetAccessible=true -Xms4096m") \
    .enableHiveSupport() \
    .getOrCreate()

spark.conf.set("spark.sql.sources.partitionOverwriteMode","dynamic")

def camel_case(s):
  s = sub(r"(_|-)+", " ", s).title().replace(" ", "")
  return ''.join([s[0].lower(), s[1:]])

def spark_df_to_json_list(df):
    df_collect=df.collect()

    json_list = []
    for row in df_collect:
        d = {}
        for col in df.schema.names:
            val = row[col]
            if 'null' == val:
                val = None
            d[camel_case(col)]=val
        json_list.append(d)

    print(json_list[0])
    return json_list

def divide_chunks(list_data, n):
    # looping till length l
    for i in range(0, len(list_data), n):
        yield list_data[i:i + n]

def post_json(url = "", data = []):
    url = "http://192.168.128.96:6000/marketing_platform/adReportLevelDataTransfer/" + url
    r = requests.post(url, json=data)
    print(r.status_code, r.text)
    return r.status_code


def transferRecord():
    df=spark.read.option("header",'true').csv("csv/ad_report_mark_record.all0.csv")
    df.printSchema()
    json_list = spark_df_to_json_list(df)
    print("length of data:", len(json_list))
    for sub_list in divide_chunks(json_list, 500):
        post_json("transferRecord", sub_list)
        
def transferLevel():
    df=spark.read.option("header",'true').csv("csv/ad_report_mark_level.csv")
    df.printSchema()
    json_list = spark_df_to_json_list(df)
    print("length of data:", len(json_list))
    for sub_list in divide_chunks(json_list, 500):
        post_json("transferLevel", sub_list)
        
def transferForbid():
    df=spark.read.option("header",'true').csv("csv/ad_report_mark_forbid.csv")
    df.printSchema()
    json_list = spark_df_to_json_list(df)
    print("length of data:", len(json_list))
    for sub_list in divide_chunks(json_list, 500):
        post_json("transferForbid", sub_list)
        
# transferForbid()
```
